```python
# File to Load (Remember to Change These)
this_year = "Resources/JC-202001-citibike-tripdata.csv"
past_year = "Resources/JC-202012-citibike-tripdata.csv"


# Read School and Student Data File and store into Pandas DataFrames
present_year = pd.read_csv(this_year)
year_ending = pd.read_csv(past_year)

#year_complete = pd.merge(this_year, past_year, how="left", on=["start_station_id", "start_station_id"])
#year_complete.head()
```

```python
present_year['stoptime']=pd.to_datetime(present_year['stoptime'])
display(present_year)
```

```python
present_year['tripduration']=pd.to_datetime(present_year['tripduration'], unit='m')
display(present_year)
```

```python
#present_year['start station name'].astype('str')

present_year["start station name"] = present_year['start station name'].astype('str')
present_year.dtypes
```

```python
present_year.info()
```

```python
present_year['gender'].replace(to_replace=[0, 1, 2], value=['Unknown', 'Male', 'Female'],inplace=True)
present_year
```

```python
presentyear = datetime.now().year

present_year["Age"] = presentyear - present_year["birth year"]

present_year.head()
```
